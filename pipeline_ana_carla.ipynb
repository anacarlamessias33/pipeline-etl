## Importações

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Extração

```python
caminho_arquivo = '/home/anacarla/Downloads/archive/fipe_cars.csv'
```

```python
dados = pd.read_csv(caminho_arquivo)
```

```python
dados.head(10)
```

## Transformação

```python
dados['avg_price_dolar'] = dados['avg_price_brl'].div(5.06)
```

```python
dados.head(10)
```

## Carregamento

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.barplot(data=dados, x='brand', y='avg_price_dolar', estimator='mean')
ax.tick_params(axis='x', labelrotation=90)
ax.set_title('Preço médios do carros por fabricante')
ax.set_ylabel('Preço médio em dólar')
ax.set_xlabel('Fabricante')

fig.savefig('avg_price_by_brand.png')
```
